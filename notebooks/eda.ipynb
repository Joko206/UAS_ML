# Exploratory Data Analysis (EDA)

Analisis ini merupakan bagian dari Proyek Analisis Prediktif.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load data
df = pd.read_csv('../data/heart.csv')
df.head()
```

### 1. Pemeriksaan Data yang Hilang (Missing Values)

```python
missing_values = df.isnull().sum()
print(missing_values)
print("\nTotal data hilang:", missing_values.sum())
```

### 2. Analisis Distribusi Data
Melihat distribusi kelas pada target (Penyakit Jantung) dan beberapa fitur numerik.

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='HeartDisease', data=df)
plt.title('Distribusi Kelas Penyakit Jantung (0=Normal, 1=Disease)')
plt.show()
```

```python
numerical_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
df[numerical_cols].hist(bins=15, figsize=(15, 10), layout=(2, 3))
plt.suptitle('Distribusi Fitur Numerik')
plt.show()
```

### 3. Identifikasi Pencilan (Outliers)
Menggunakan boxplot untuk mendeteksi anomali/pencilan pada data numerik.

```python
plt.figure(figsize=(15, 8))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i+1)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

### 4. Korelasi Antar-Fitur
Melihat hubungan linear antara fitur numerik.

```python
plt.figure(figsize=(8, 6))
correlation = df[numerical_cols + ['HeartDisease']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Korelasi Fitur Numerik')
plt.show()
```
